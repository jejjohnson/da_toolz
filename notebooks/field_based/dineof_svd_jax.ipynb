```python
import equinox as eqx
class PCAJAX(eqx.Module):
    n_features: int = eqx.field(static=True)
    n_components: int = eqx.field(static=True)
    n_iterations: int = eqx.field(static=True)
    size: int = eqx.field(static=True)
    Q: Array = eqx.field(static=True)

    def __init__(self, n_features: int, n_components: int=10, n_iterations: int=5, rng=jax.random.PRNGKey(123)):
        self.n_features = n_features
        self.n_components = n_components
        self.n_iterations = n_iterations
        self.size = jnp.minimum(2 * self.n_components, self.n_features)
        self.Q = jax.random.normal(rng, shape=(self.n_features, self.size))
        

    def reconstruct(self, X: Array) -> Array:

        U, S, VT = self.rsvd(X)
        # print(X.shape, U.shape, S.shape, VT.shape)
        # Z = X @ VT.T
        # return Z @ VT

        Z = einx.dot("N D, d D -> N d", X, VT)
        # print(Z.shape, VT.shape)

        X_recon = einx.dot("N d, d D -> N D", Z, VT)
        # print(X_recon.shape)
        return X_recon

    def rsvd(self, x):

        # remove means
        means = jnp.mean(x, axis=0, keepdims=True)
        x = x - means
    
        # Generate n_features normal vectors of the given size
        def step_fn(q, _):
            q, _ = jax.scipy.linalg.lu(x @ q, permute_l=True)
            q, _ = jax.scipy.linalg.lu(x.T @ q, permute_l=True)
            return q, None
    
        Q, _ = jax.lax.scan(step_fn, init=self.Q, xs=None, length=self.n_iterations)
        Q, _ = jax.scipy.linalg.qr(x @ Q, mode="economic")
        B = Q.T @ x
    
        U, S, VT = jax.scipy.linalg.svd(B, full_matrices=False)
    
        return U, S, VT
```
